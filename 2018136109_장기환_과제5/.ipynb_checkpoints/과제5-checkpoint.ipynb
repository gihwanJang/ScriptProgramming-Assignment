ë¬¸ì œ : ë‹¤ìŒ ê° ìš”êµ¬ì‚¬í•­ ëª¨ë‘ë¥¼ ë§Œì¡±ì‹œí‚¤ëŠ” Counter í´ë˜ìŠ¤ë¥¼ ì½”ë”©í•˜ì‹œì˜¤ (ì •ë‹µì„ ê° ìš”êµ¬ì‚¬í•­ë³„ë¡œ ì…ë ¥í•  í•„ìš” ì—†ì´ 3ë²ˆ ë¬¸ì œì— ëŒ€í•´ 1ê°œì˜ í´ë˜ìŠ¤ ì •ì˜ ì½”ë“œë¥¼ ì œì‹œí•˜ë©´ ëœë‹¤.)
<pre>&#176; ìš”êµ¬ì‚¬í•­ 1. ìƒì„±ìì— count ê°’ê³¼ step ê°’ì„ ì¸ìë¡œ ë°›ì„ ìˆ˜ ìˆë‹¤.  
    &squarf; count: Counter ì¸ìŠ¤í„´ìŠ¤ê°€ ì§€ë‹ˆëŠ” ì´ˆê¸° ì •ìˆ˜ ê°’  
    &squarf; step: Counter ì¸ìŠ¤í„´ìŠ¤ì˜ countê°€ ì¦ê°€ë˜ëŠ” ì¦ë¶„ (defalt ê°’: 1) 
            >>>c = Counter(10)

&#176; ìš”êµ¬ì‚¬í•­ 2. ë‹¤ìŒê³¼ ê°™ì´ Counterì˜ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ì¶œë ¥ì„ í•´ì£¼ëŠ” __str__() ë©”ì†Œë“œë¥¼ Counter í´ë˜ìŠ¤ ë‚´ì— êµ¬í˜„í•˜ì‹œì˜¤.  
            >>> print(c)

&#176; ìš”êµ¬ì‚¬í•­ 3. ë‹¤ìŒê³¼ ê°™ì´ stepì— ì£¼ì–´ì§„ ì¦ë¶„ë§Œí¼ countë¥¼ ì¦ê°€ì‹œí‚¤ëŠ” incr() ë©”ì†Œë“œë¥¼ Counter í´ë˜ìŠ¤ ë‚´ì— êµ¬í˜„í•˜ì‹œì˜¤.   
            >>> c.incr()

&#176; ìš”êµ¬ì‚¬í•­ 4. Counter í´ë˜ìŠ¤ ë‚´ì— ê´€ë ¨ ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì—¬ ì¸ìŠ¤í„´ìŠ¤ ê°ì²´ë¥¼ ì§ì ‘ í˜¸ì¶œ(call)í•  ìˆ˜ ìˆë„ë¡ í•˜ì‹œì˜¤. ì¸ìŠ¤í„´ìŠ¤ ê°ì²´ë¥¼ ì§ì ‘ í˜¸ì¶œí–ˆì„ ë•Œì— ë‚´ë¶€ì ìœ¼ë¡œ incr() ë©”ì†Œë“œë¥¼ í˜¸ì¶œí•˜ëŠ” ë°©ë²•ìœ¼ë¡œ êµ¬í˜„í•˜ì‹œì˜¤.

&#176; ìš”êµ¬ì‚¬í•­ 5. ë‹¤ìŒê³¼ ê°™ì€ ë‘ ê°œì˜ ì‚°ìˆ  ì—°ì‚° (+, -)ì´ ìˆ˜í–‰ë  ìˆ˜ ìˆë„ë¡ Counter í´ë˜ìŠ¤ ë‚´ì— ê´€ë ¨ ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì‹œì˜¤.
            >>> c = c + 5 
            >>> d = d - 5 
            >>> print(c) 
            [Count (step: 1)] 17 
            >>> print(d) 
            [Count (step: 2)] 9 
&#176; ìš”êµ¬ì‚¬í•­ 6. ë‹¤ìŒê³¼ ê°™ì€ ê´€ê³„ì—°ì‚° (>, &lt;, ==)ì´ ìˆ˜í–‰ë  ìˆ˜ ìˆë„ë¡ Counter í´ë˜ìŠ¤ ë‚´ì— ê´€ë ¨ ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì‹œì˜¤.
            >>> print(c > 10) 
            True 
            >>> print(d >= 10) 
            False 
            >>> print(c &lt; 10) 
            False 
            >>> print(d &lt;= 10) 
            True 
            >>> print(c == 17) 
            True 
            >>> print(d != 9) 
            False 
                   </pre>

```python
class Counter:
    #countì™€ stepì„ ë°›ëŠ” ìƒì„±ì(stepì€ default ê°’ìœ¼ë¡œ 1ì„ ê°€ì§)
    def __init__(self, count, step=1):
        self.count=count
        self.step=step
    #ì¸ìŠ¤í„´ìŠ¤ë¥¼ ì¶œë ¥
    def __str__(self):
        return '[Count (step: {})] {}'.format(self.step,self.count)
    #countë¥¼ stepí¬ê¸°ë§Œí¼ ì¦ê°€ì‹œí‚¤ëŠ” ë©”ì†Œë“œ
    def incr(self):
        self.count+=self.step
    #í˜¸ì¶œì‹œ incrë©”ì†Œë“œ ì‹¤í–‰
    def __call__(self):
        self.incr()
    #+ì—°ì‚°ì ì¬ì •ì˜(count ê°’ +ì—°ì‚°)
    def __add__(self, other):
        self.count+=other
        return self
    #-ì—°ì‚°ì ì¬ì •ì˜(count ê°’ -ì—°ì‚°)
    def __sub__(self, other):
        self.count-=other
        return self
    #<ì—°ì‚°ì ì¬ì •ì˜(count ê°’ <ì—°ì‚°)
    def __lt__(self, other):
        return self.count<other
    #>ì—°ì‚°ì ì¬ì •ì˜(count ê°’ >ì—°ì‚°)
    def __gt__(self, other):
        return self.count>other
    #<=ì—°ì‚°ì ì¬ì •ì˜(count ê°’ <=ì—°ì‚°)
    def __le__(self, other):
        return self.count<=other
    #>=ì—°ì‚°ì ì¬ì •ì˜(count ê°’ >=ì—°ì‚°)
    def __ge__(self, other):
        return self.count>=other
    #==ì—°ì‚°ì ì¬ì •ì˜(count ê°’ ==ì—°ì‚°)
    def __eq__(self, other):
        return self.count==other
    #!=ì—°ì‚°ì ì¬ì •ì˜(count ê°’ !=ì—°ì‚°)
    def __ne__(self, other):
        return self.count!=other
```

```python
#ìš”êµ¬ì‚¬í•­ 1
c=Counter(10)
d=Counter(10,2)
```

```python
#ìš”êµ¬ì‚¬í•­ 2
print(c)
print(d)
```

```python
#ìš”êµ¬ì‚¬í•­ 3
c.incr()
d.incr()
print(c)
print(d)
```

```python
#ìš”êµ¬ì‚¬í•­ 4
c()
d()
print(c)
print(d)
```

```python
#ìš”êµ¬ì‚¬í•­ 5
c=c+5
d=d-5
print(c)
print(d)
```

```python
#ìš”êµ¬ì‚¬í•­ 6
print(c>10)
print(d>=10)
print(c<10)
print(d<=10)
print(c==17)
print(d!=9)
```

## <span style="color:#87CEEB">ê³¼ì œ 5.4</span>
ë¬¸ì œ : ë‹¤ìŒì€ ë‚´ì¥ ìë£Œí˜• listë¥¼ ì„œë¸Œí´ë˜ì‹±í•˜ì—¬ ë§Œë“  MySet í´ë˜ìŠ¤ ì •ì˜ ë‚´ìš©ì´ë‹¤. ë‹¤ìŒ í´ë˜ìŠ¤ ì •ì˜ì—ì„œ &#95;&#95;init&#95;&#95;(), &#95;&#95;str()&#95;&#95;(), elimicate_duplicate()ì˜ ì„¸ ê°œì˜ ë©”ì†Œë“œ ì½”ë“œ ë‚´ìš©ì„ ìì‹ ì´ ë‹¤ë¥¸ ì‚¬ëŒì—ê²Œ ê°€ë¥´ì¹œë‹¤ê³  ìƒê°í•˜ë©° ì„¤ëª…í•´ë³´ì‹œì˜¤.  
<pre>&#176; MySetì€ ì§‘í•©(Set) ìë£Œí˜•ì„ ì •ì˜í•˜ë ¤ëŠ” ì˜ë„í•˜ì— ë§Œë“¤ì–´ì§„ í´ë˜ìŠ¤ì´ë‹¤.  <pre>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

## <span style="color:#87CEEB">ê³¼ì œ 5.5</span>
ë¬¸ì œ : 4ë²ˆ ë¬¸ì œì— ì •ì˜ëœ MySet í´ë˜ìŠ¤ì— ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì—¬ ë‹¤ìŒ ê° ìš”êµ¬ì‚¬í•­ ëª¨ë‘ë¥¼ ë§Œì¡±ì‹œí‚¤ëŠ” ì½”ë”©ì„ ì œì‹œí•˜ì‹œì˜¤.  
<pre>&#176; ì •ë‹µì„ ê° ìš”êµ¬ì‚¬í•­ë³„ë¡œ ì…ë ¥í•  í•„ìš” ì—†ì´ ìš”êµ¬ì‚¬í•­ 3ê°œ ì „ì²´ì— ëŒ€í•´ 1ê°œì˜ MySet í´ë˜ìŠ¤ ì •ì˜ ì½”ë“œë¥¼ ì œì‹œí•˜ê³  ì£¼ì„ë“±ìœ¼ë¡œ í•´ë‹¹ ì½”ë“œë¥¼ ì„¤ëª…í•œë‹¤.  
&#176; ìš”êµ¬ì‚¬í•­ 1. | ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ í•©ì§‘í•©ì„ ë°˜í™˜í•œë‹¤.
            >>> u = s | t 
            >>> print(u) 
            MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
&#176; ìš”êµ¬ì‚¬í•­ 2. &amp; ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ êµì§‘í•©ì„ ë°˜í™˜í•œë‹¤.
            >>> u = s &amp; t 
            >>> print(u) 
            MySet: {2, 3} 
&#176; ìš”êµ¬ì‚¬í•­ 3. - ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ ì°¨ì§‘í•©ì„ ë°˜í™˜í•œë‹¤. 
            >>> s = MySet([1, 2, 3]) 
            >>> t = MySet([3, 4, 5]) 
            >>> u = s - t 
            >>> print(u) 
            MySet: {1, 2}  
<pre>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
        
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    # |ì—°ì‚°ì ì¬ì •ì˜(ì„ì˜ ë°°ì—´ì— ë‘ì¸ìì˜ ê°’ë“¤ì„ appendí›„ í•´ë‹¹ ë°°ì—´ë¡œ MySetìƒì„±)
    def __or__(self, other):
        result = []
        for e in self:
            result.append(e)
        for e in other:
            result.append(e)
        return MySet(result)
    # &ì—°ì‚°ì ì¬ì •ì˜(ì„ì˜ ë°°ì—´ì— ë‘ì¸ìì˜ ê²¹ì¹˜ëŠ” ê°’ë“¤ì„ appendí›„ í•´ë‹¹ ë°°ì—´ë¡œ MySetìƒì„±)
    def __and__(self, other):
        result = []
        for e in self:
            if e in other:
                result.append(e)
        return MySet(result)
    # -ì—°ì‚°ì ì¬ì •ì˜(ì„ì˜ ë°°ì—´ì— ë‘ì¸ìì˜ ê²¹ì¹˜ì§€ ì•ŠëŠ” ê°’ë“¤ì„ appendí›„ í•´ë‹¹ ë°°ì—´ë¡œ MySetìƒì„±)
    def __sub__(self, other):
        result = []
        for e in self:
            if e not in other:
                result.append(e)
        return MySet(result)
```

```python
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
u = s | t 
v = s & t
w = s - t
print(s)
print(t)
print(u)
print(v)
print(w)
```

```python
import requests
import re
import html
import string

class SearchEngine:
    #ë¶ˆìš©ì–´ ë°°ì—´ ìƒì„±
    stopwords = []
    #ìƒì„±ì ì¬ì •ì˜(ë¶ˆìš©ì–´ íŒŒì¼ìƒì„±, url, urlë‹¨ì–´ ì…ë ¥)
    def __init__(self, *args):
        SearchEngine.stopWordsMake()
        self.urls = []
        self.urlsDict = {}
        for url in args:
            self.addUrl(url)
    #url ì¶”ê°€ë©”ì†Œë“œ(url, urlë‹¨ì–´ ë°°ì—´ì— ì¶”ê°€)
    def addUrl(self, url):
        self.urls.append(url)
        self.urlsDict[url] = self.countWords(url)
    #url ì‚­ì œë©”ì†Œë“œ(url, urlë‹¨ì–´ ë°°ì—´ì—ì„œ ì‚­ì œ)
    def removeUrl(self, url):
        if url in self.urls:
            self.urls.remove(url)
            del self. urlsDict[url]
    #urlëª©ë¡ ì¶œë ¥ ë©”ì†Œë“œ
    def listUrls(self):
        for url in self.urls:
            print(url)
    #urlë“¤ì˜ ë‹¨ì–´ë“¤ì„ ì¢…í•©í•˜ì—¬ ì‚¬ì „í˜• ì¶œë ¥(urlì´ ì—†ìœ¼ë©´ none)
    def getWordsFrequency(self):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        return urlDict
    #urlë“¤ì˜ ë‹¨ì–´ë“¤ì„ ì¢…í•©í•˜ì—¬ ìµœë‹¤ë¹ˆë„ ë‹¨ì–´ ì‚¬ì „í˜• ì¶œë ¥(urlì´ ì—†ìœ¼ë©´ none)
    def getMaxFreqencyWords(self):
        if len(self.urls)==0:
            return None
        keywords = list(self.getWordsFrequency().items())
        keywords.sort(key=lambda k: k[1], reverse=True)
        s={}
        while True:
            i=0
            s[keywords[i][0]]=keywords[i][1]
            ++i
            if(keywords[i-1][1]!=keywords[i][0]):
                break
        return s
    #ê²€ìƒ‰ ë©”ì†Œë“œ(ê° urlë“¤ì˜ ë‹¨ì–´ë“¤ê³¼ ë¹„êµí•˜ì—¬ ìœ ì‚¬ë„ ìˆœìœ¼ë¡œ ì¶œë ¥)
    def searchUrlByWord(self, suchKey):
        suchKey = suchKey.split()
        suchKey = SearchEngine.wordsFilter(suchKey)
        ranking = self.similarity(suchKey)
        return sorted(ranking.items(), reverse=True, key=lambda x: x[1])
    #urlë“¤ì˜ htmltxtë¥¼ ë°›ì•„ì™€ ë‹¨ì–´ì™€ ê°¯ìˆ˜ë¥¼ ì¶”ì¶œí•´ ì‚¬ì „í˜• ë°˜í™˜
    def countWords(self, url):
        req = requests.get(url)
        source = re.sub(r'\n', ' ', req.text)
        source = re.sub(r'<(script|style).*?</\1>', '', source)
        source = re.sub(r'<!(--)?.*?-->', '', source)
        source = re.sub(r'</?.*?>', '', source)
        words=[]   
        source=html.unescape(source)

        words = source.split()
        words = SearchEngine.wordsFilter(words)
        word1 = SearchEngine.numWord(words)
        word2 = SearchEngine.count(word1,words)
    
        dictionary = dict(zip(word1, word2))
        return dictionary
    #urlë“¤ì˜ ë‹¨ì–´ë“¤ê³¼ ì…ë ¥ ë‹¨ì–´ ê°„ì˜ ìœ ì‚¬ë„ ê³„ì‚°
    def similarity(self,words):
        dic={}
        for key in self.urlsDict.keys():
            count = 0
            sumWords = 0
            for k,v in self.urlsDict[key].items():
                for w in words:
                    if w == k:
                        count += v
                sumWords += v
            dic[key]=(count*100)/sumWords
        return dic
    #ë¶ˆìš©ì–´ ë°°ì—´ ì…ë ¥
    @staticmethod
    def stopWordsMake():
        req_stopwords = requests.get('https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt')
        req_stopwords2 = requests.get("https://gist.githubusercontent.com/sebleier/554280/raw/7e0e4a1ce04c2bb7bd41089c9821dbcf6d0c786c/NLTK's%2520list%2520of%2520english%2520stopwords")
        source_stopwords = req_stopwords.text
        source_stopwords2 = req_stopwords2.text
        SearchEngine.stopwords = source_stopwords.split()+source_stopwords2.split()
    #ë‹¨ì–´ë“¤ ì¤‘ë³µ ì œê±°
    @staticmethod
    def numWord(words):
        a=[]
        for i in words:
            if(not i in a):
                a.append(i)
        return a
    @staticmethod  
    #íŠ¹ìˆ˜ ë¬¸ì,ë¶ˆìš©ì–´ ì œê±°
    def wordsFilter(words):
        words = [i for i in words if (i not in string.punctuation) and (i.lower() not in SearchEngine.stopwords)]
        return words
    #ë‹¨ì–´ ì¤‘ë³µ ê°¯ìˆ˜ ì¹´ìš´íŠ¸
    @staticmethod    
    def count(word1,word2):
        a=[]
        for i in word1:
            count=0
            for j in word2:
                if i==j:
                    count+=1
            a.append(count)
        return a
```

## <span style="color:#87CEEB">ê³¼ì œ 5.8</span>
ë¬¸ì œ :  ìœ„ 7ë²ˆ ë¬¸ì œì—ì„œ ì •ì˜í•œ SearchEngine í´ë˜ìŠ¤ë¥¼ ìƒì†í•˜ì—¬ SearchEngineWithOrderedWebWords í´ë˜ìŠ¤ë¥¼ ì •ì˜í•˜ê³  ìŠˆí¼í´ë˜ìŠ¤ì— ì •ì˜ëœ getWordsFrequency() ë©”ì†Œë“œë¥¼ ì˜¤ë²„ë¼ì´ë“œí•˜ì—¬ ë‹¨ì–´ ì¶œí˜„ ë¹ˆë„ë¥¼ ë‚´ë¦¼ ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ì¶œë ¥í•˜ì‹œì˜¤.
<pre>&#176; ë¦¬ìŠ¤íŠ¸ ë‚´ì˜ ê° ì›ì†ŒëŠ” ë‹¨ì–´ì™€ ë¹ˆë„ë¥¼ ìŒìœ¼ë¡œ ì§€ë‹ˆëŠ” íŠœí”Œì´ë‹¤.
&#176; getWordFrequence() ë©”ì†Œë“œì— reverse ë¼ëŠ” ì¸ìë¥¼ ë§Œë“¤ê³  true ë˜ëŠ” falseë¥¼ ì¸ìë¡œ ë°›ì„ ìˆ˜ ìˆë„ë¡ í•œë‹¤.
    &squarf; reverse ì¸ìì˜ ë””í´íŠ¸ ê°’ì€ false ì´ë©°, ê¸°ë³¸ì ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤.
    &squarf; reverse ì¸ìì— trueë¥¼ ë„£ìœ¼ë©´ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤.
            >>> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

            >>> w4.getWordsFrequency()
            [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

            >>> w4.getWordsFrequency(reverse=true)
            [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]
        <pre>

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    # getWordsFrequencyë©”ì†Œë“œ ì¬ì •ì˜(ê¸°ì¡´ ë°˜í™˜ ì‚¬ì „ ì •ë ¬ì¸ìê°€ Falseì´ë©´ ë‚´ë¦¼ Trueì´ë©´ ì˜¤ë¦„ ì°¨ìˆœì´ë‹¤)
    def getWordsFrequency(self,rev=False):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        if(rev==False):
            return sorted(urlDict.items(), reverse=True, key=lambda x: x[1])
        elif(rev==True):
            return sorted(urlDict.items(), key=lambda x: x[1])
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
print()
print()
print()
print()
print(w4.getWordsFrequency(True))
```

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    
    def getWordsFrequency(self,rev=False):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        if(rev==False):
            return sorted(urlDict.items(), reverse=True, key=lambda x: x[1])
        elif(rev==True):
            return sorted(urlDict.items(), key=lambda x: x[1])
    #ì´í„°ë ˆì´í„° ì¬ì •ì˜ getWordsFrequencyì˜ ì‚¬ì „ì„ ë°›ì•„ì™€ í•´ë‹¹ ì‚¬ì „ì„ ì´ìš©í•˜ì—¬ ì „ì§„í•œë‹¤.
    def __iter__(self):
        wordDict=self.getWordsFrequency()
        return iter(wordDict)
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
for i in w4:
    print(i)
```
