## <span style="color:#87CEEB">과제 5.2</span>
문제 : 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

<span style="color:red">&lt; 답 &gt; </span>   
다형성(Polymorphism)
<pre>&#176; 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능
    &squarf; 연산자 오버로딩(Operator Overloding)도 다형성과 관련된 중요한 기술
        &bull; 연산자 오버로딩:a와b의 객체 형에 따라 a+b의 +연산자 행동 방식이 변경되는 것
&#176; 다형성의 장점
    &squarf; 적은 코딩으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있음
    &squarf; 일관된 코딩 방식이 유지됨
    &squarf; 프로그램 작성 코드량이 줄어든다.
    &squarf; 코드의 가독성을 높혀준다.
&#176; 파이썬에서 다형성의 장점
    &squarf; 형 선언이 없다는 점에서 파이썬에서는 다형성을 적용하기가 더욱 용이하다.
    &squarf; 실시간으로 객체의 형이 결경 => 하나의 메소드에 연관시킬 수 있는 객체의 종류에 제한이 없다.
        &bull; 즉, 상속 관계등의 클래스 간의 관계 설정 필요 없이 다형성 적용 가능
<pre>   

아래에서 Vehicle,Car,Airplane이 같은 printKind라는 멤버 함수 호출에 대해 각각 다르게 반응하고 있다.

```python
class Vehicle:
    def printKind(self):
        print("vehicle")
#Vehicle을 상속받는 클래스로 printKind메소드 재정의
class Car(Vehicle):
    def printKind(self):
        print("car")
#Vehicle을 상속받는 클래스로 printKind메소드 재정의
class Airplane(Vehicle):
    def printKind(self):
        print("airplane")
#각 클래스의 printKind실행
for each in(Vehicle(),Car(),Airplane()):
    each.printKind()
```

## <span style="color:#87CEEB">과제 5.3</span>
문제 : 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
<pre>&#176; 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.  
    &squarf; count: Counter 인스턴스가 지니는 초기 정수 값  
    &squarf; step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1) 
            >>>c = Counter(10) 
            >>>d = Counter(10, 2)  
&#176; 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.  
            >>> print(c) 
            [Count (step: 1)] 10  
            >>> print(d)  
            [Count (step: 2)] 10  
&#176; 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.   
            >>> c.incr()  
            >>> d.incr()  
            >>> print(c)  
            [Count (step: 1)] 11  
            >>> print(d)  
            [Count (step: 2)] 12  
&#176; 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.  
            >>> c()  
            >>> d()  
            >>> print(c)  
            [Count (step: 1)] 12  
            >>> print(d)  
            [Count (step: 2)] 14    
&#176; 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
            >>> c = c + 5 
            >>> d = d - 5 
            >>> print(c) 
            [Count (step: 1)] 17 
            >>> print(d) 
            [Count (step: 2)] 9 
&#176; 요구사항 6. 다음과 같은 관계연산 (>, &lt;, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
            >>> print(c > 10) 
            True 
            >>> print(d >= 10) 
            False 
            >>> print(c &lt; 10) 
            False 
            >>> print(d &lt;= 10) 
            True 
            >>> print(c == 17) 
            True 
            >>> print(d != 9) 
            False 
                   </pre>

```python
class Counter:
    #count와 step을 받는 생성자(step은 default 값으로 1을 가짐)
    def __init__(self, count, step=1):
        self.count=count
        self.step=step
    #인스턴스를 출력
    def __str__(self):
        return '[Count (step: {})] {}'.format(self.step,self.count)
    #count를 step크기만큼 증가시키는 메소드
    def incr(self):
        self.count+=self.step
    #호출시 incr메소드 실행
    def __call__(self):
        self.incr()
    #+연산자 재정의(count 값 +연산)
    def __add__(self, other):
        self.count+=other
        return self
    #-연산자 재정의(count 값 -연산)
    def __sub__(self, other):
        self.count-=other
        return self
    #<연산자 재정의(count 값 <연산)
    def __lt__(self, other):
        return self.count<other
    #>연산자 재정의(count 값 >연산)
    def __gt__(self, other):
        return self.count>other
    #<=연산자 재정의(count 값 <=연산)
    def __le__(self, other):
        return self.count<=other
    #>=연산자 재정의(count 값 >=연산)
    def __ge__(self, other):
        return self.count>=other
    #==연산자 재정의(count 값 ==연산)
    def __eq__(self, other):
        return self.count==other
    #!=연산자 재정의(count 값 !=연산)
    def __ne__(self, other):
        return self.count!=other
```

```python
#요구사항 1
c=Counter(10)
d=Counter(10,2)
```

```python
#요구사항 2
print(c)
print(d)
```

```python
#요구사항 3
c.incr()
d.incr()
print(c)
print(d)
```

```python
#요구사항 4
c()
d()
print(c)
print(d)
```

```python
#요구사항 5
c=c+5
d=d-5
print(c)
print(d)
```

```python
#요구사항 6
print(c>10)
print(d>=10)
print(c<10)
print(d<=10)
print(c==17)
print(d!=9)
```

## <span style="color:#87CEEB">과제 5.4</span>
문제 : 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 &#95;&#95;init&#95;&#95;(), &#95;&#95;str()&#95;&#95;(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.  
<pre>&#176; MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.  <pre>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

<span style="color:red">&lt; 답 &gt; </span>   
<pre>&#176; elimicate_duplicate()
    &squarf; static 메소드로 정의 되어 일반메소드 처럼 MySet.eliminate_duplicate()롤 사용할 수 있다.
    &squarf; 해당 메소드는 리스트를 인자로 받아 해당 리스트의 값들 중 중복되는 값을 제거하여 주는 메소드이다.
        &bull; for문을 통해 받은 리스트의 값을 다른 리스트에 옮긴다.
        &bull; 이때 if문을 통해 옮기는 리스트에 값이 있으면 옮기지 않고 없으면 옮긴다.
        &bull; 입력받은 리스트를 빈 리스트로 초기화 한다.
        &bull; for문을 통해 다시 옮긴 리스트를 빈리스트로 값을 옮긴다.        
&#176; &#95;&#95;init&#95;&#95;()
    &squarf; 생성자 메소드를 재정의 한 것이다.
        &bull; for문을 통해 받은 리스트의 값을 자신의 리스트에 옮긴다.
        &bull; MySet.eliminate_duplicate(self)를 통해 중복된 값을 삭제 한다.
&#176; &#95;&#95;str&#95;&#95;()    
    &squarf; 객체 출력 메소드를 재정의 한 것이다.
        &bull; "MySet: {"라는 시작 문구로 설정한다.
        &bull; 배열의 인자를 문자열+연산자를 통해 추가시키고 해당+연산자를 통해 " ,"로 구분시킨다.
        &bull; 해당 문자열의 마지막 구분자 즉 -2번째에 "}" +연사자를 통해 추가후 return 한다.

## <span style="color:#87CEEB">과제 5.5</span>
문제 : 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.  
<pre>&#176; 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
&#176; 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
            >>> u = s | t 
            >>> print(u) 
            MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
&#176; 요구사항 2. &amp; 연산으로 두 집합의 교집합을 반환한다.
            >>> u = s &amp; t 
            >>> print(u) 
            MySet: {2, 3} 
&#176; 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다. 
            >>> s = MySet([1, 2, 3]) 
            >>> t = MySet([3, 4, 5]) 
            >>> u = s - t 
            >>> print(u) 
            MySet: {1, 2}  
<pre>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
        
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    # |연산자 재정의(임의 배열에 두인자의 값들을 append후 해당 배열로 MySet생성)
    def __or__(self, other):
        result = []
        for e in self:
            result.append(e)
        for e in other:
            result.append(e)
        return MySet(result)
    # &연산자 재정의(임의 배열에 두인자의 겹치는 값들을 append후 해당 배열로 MySet생성)
    def __and__(self, other):
        result = []
        for e in self:
            if e in other:
                result.append(e)
        return MySet(result)
    # -연산자 재정의(임의 배열에 두인자의 겹치지 않는 값들을 append후 해당 배열로 MySet생성)
    def __sub__(self, other):
        result = []
        for e in self:
            if e not in other:
                result.append(e)
        return MySet(result)
```

```python
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
u = s | t 
v = s & t
w = s - t
print(s)
print(t)
print(u)
print(v)
print(w)
```

    &squarf; 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오. 
</pre>

```python
import requests
import re
import html
import string

class SearchEngine:
    #불용어 배열 생성
    stopwords = []
    #생성자 재정의(불용어 파일생성, url, url단어 입력)
    def __init__(self, *args):
        SearchEngine.stopWordsMake()
        self.urls = []
        self.urlsDict = {}
        for url in args:
            self.addUrl(url)
    #url 추가메소드(url, url단어 배열에 추가)
    def addUrl(self, url):
        self.urls.append(url)
        self.urlsDict[url] = self.countWords(url)
    #url 삭제메소드(url, url단어 배열에서 삭제)
    def removeUrl(self, url):
        if url in self.urls:
            self.urls.remove(url)
            del self. urlsDict[url]
    #url목록 출력 메소드
    def listUrls(self):
        for url in self.urls:
            print(url)
    #url들의 단어들을 종합하여 사전형 출력(url이 없으면 none)
    def getWordsFrequency(self):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        return urlDict
    #url들의 단어들을 종합하여 최다빈도 단어 사전형 출력(url이 없으면 none)
    def getMaxFreqencyWords(self):
        if len(self.urls)==0:
            return None
        keywords = list(self.getWordsFrequency().items())
        keywords.sort(key=lambda k: k[1], reverse=True)
        s={}
        while True:
            i=0
            s[keywords[i][0]]=keywords[i][1]
            ++i
            if(keywords[i-1][1]!=keywords[i][0]):
                break
        return s
    #검색 메소드(각 url들의 단어들과 비교하여 유사도 순으로 출력)
    def searchUrlByWord(self, suchKey):
        suchKey = suchKey.split()
        suchKey = SearchEngine.wordsFilter(suchKey)
        ranking = self.similarity(suchKey)
        return sorted(ranking.items(), reverse=True, key=lambda x: x[1])
    #url들의 htmltxt를 받아와 단어와 갯수를 추출해 사전형 반환
    def countWords(self, url):
        req = requests.get(url)
        source = re.sub(r'\n', ' ', req.text)
        source = re.sub(r'<(script|style).*?</\1>', '', source)
        source = re.sub(r'<!(--)?.*?-->', '', source)
        source = re.sub(r'</?.*?>', '', source)
        words=[]   
        source=html.unescape(source)

        words = source.split()
        words = SearchEngine.wordsFilter(words)
        word1 = SearchEngine.numWord(words)
        word2 = SearchEngine.count(word1,words)
    
        dictionary = dict(zip(word1, word2))
        return dictionary
    #url들의 단어들과 입력 단어 간의 유사도 계산
    def similarity(self,words):
        dic={}
        for key in self.urlsDict.keys():
            count = 0
            sumWords = 0
            for k,v in self.urlsDict[key].items():
                for w in words:
                    if w == k:
                        count += v
                sumWords += v
            dic[key]=(count*100)/sumWords
        return dic
    #불용어 배열 입력
    @staticmethod
    def stopWordsMake():
        req_stopwords = requests.get('https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt')
        req_stopwords2 = requests.get("https://gist.githubusercontent.com/sebleier/554280/raw/7e0e4a1ce04c2bb7bd41089c9821dbcf6d0c786c/NLTK's%2520list%2520of%2520english%2520stopwords")
        source_stopwords = req_stopwords.text
        source_stopwords2 = req_stopwords2.text
        SearchEngine.stopwords = source_stopwords.split()+source_stopwords2.split()
    #단어들 중복 제거
    @staticmethod
    def numWord(words):
        a=[]
        for i in words:
            if(not i in a):
                a.append(i)
        return a
    @staticmethod  
    #특수 문자,불용어 제거
    def wordsFilter(words):
        words = [i for i in words if (i not in string.punctuation) and (i.lower() not in SearchEngine.stopwords)]
        return words
    #단어 중복 갯수 카운트
    @staticmethod    
    def count(word1,word2):
        a=[]
        for i in word1:
            count=0
            for j in word2:
                if i==j:
                    count+=1
            a.append(count)
        return a
```

```python
w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com') 
w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com') 
w3 = SearchEngine()
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')
w1.listUrls()
print()
print(w1.getWordsFrequency())
print()
print(w1.getMaxFreqencyWords())
print()
print(w2.getMaxFreqencyWords())
print()
print(w1.searchUrlByWord("New"))
print()
print(SearchEngine.stopwords)
```

## <span style="color:#87CEEB">과제 5.8</span>
문제 :  위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
<pre>&#176; 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
&#176; getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
    &squarf; reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    &squarf; reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
            >>> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

            >>> w4.getWordsFrequency()
            [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

            >>> w4.getWordsFrequency(reverse=true)
            [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]
        <pre>

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    # getWordsFrequency메소드 재정의(기존 반환 사전 정렬인자가 False이면 내림 True이면 오름 차순이다)
    def getWordsFrequency(self,rev=False):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        if(rev==False):
            return sorted(urlDict.items(), reverse=True, key=lambda x: x[1])
        elif(rev==True):
            return sorted(urlDict.items(), key=lambda x: x[1])
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
print()
print()
print()
print()
print(w4.getWordsFrequency(True))
```

## <span style="color:#87CEEB">과제 5.9</span>
문제 : 다음과 같은 코딩이 가능하도록 SearchEngineWithOrderedWebWords안에 반복자와 관련된 메소드를 추가하시오.
<pre>
            >>> for i in w4:
            >>>     print(i)
            ('site, 12)
            ('science', 11)
            ('hello', 8)
            ('world', 2)
            ('program', 1)
            ('python', 1)
        <pre>

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    
    def getWordsFrequency(self,rev=False):
        if len(self.urls)==0:
            return None
        urlDict = {}
        for val in self.urlsDict.values():
            for k, v in val.items():
                if k in urlDict:
                    urlDict[k] += v
                else:
                    urlDict[k] = v
        if(rev==False):
            return sorted(urlDict.items(), reverse=True, key=lambda x: x[1])
        elif(rev==True):
            return sorted(urlDict.items(), key=lambda x: x[1])
    #이터레이터 재정의 getWordsFrequency의 사전을 받아와 해당 사전을 이용하여 전진한다.
    def __iter__(self):
        wordDict=self.getWordsFrequency()
        return iter(wordDict)
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
for i in w4:
    print(i)
```
