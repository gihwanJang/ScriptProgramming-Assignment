<pre>&#176; 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
     &starf; >>> addall([1])
       1
     &starf; >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
       45
&#176; 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오)
     &starf; >>> addallodd([1])
       1
     &starf; >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
       25
&#176;[참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
&#176;[참고 2]: sum() 등의 내장함수 사용 가능함</pre>

```python
def addall(a):
    print(sum([x for x in a]))
```

```python
def addallodd(a):
    print(sum([x for x in a if x%2==1]))
```

```python
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

문제 : 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
<pre>&#176; length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
&#176; Strange라는 숫자를 아래과 같이 정의한다.
    &starf; 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    &starf; length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        &#10026; X는 length(X)로 정확히 나누어 떨어진다.
        &#10026; X / length(X) 가 strnage이다.
&#176; 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
&#176; isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.
&#176; [예시]
        1. num = 0                      =>    True
        2. num = 56                     =>    True
        3. num = 100                    =>    False
        4. num = 7882602506492313600    =>    True
        5. num = 6012302549522098651    =>    False</pre>

```python
def isStrange(num):
    length=len(str(num))
    if(type(num)==int and (length==1 or num%length==0)):
        return True
    else:
        if(type(num)==int):
            isStrange(num/length)
    return False
```

```python
print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

## <span style="color:#87CEEB">과제 3.9</span>
문제 : 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
<pre>&#176; 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
  하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
  하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
  또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
  마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.
&#176; 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
&#176; [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!
&#176; [예시]
        1. a = [1, 2, 1, 1, 1, 2, 1, 3]    =>    [3, 4, 6, 12]
        2. a = [1, 1, 1, 1]                =>    [1, 2, 4]
        3. a = [9999] * 10**5              =>    
        [9999,19998,39996,49995,79992,99990,159984,199980,249975,319968,399960,499950,799920,999900,1249875,1599840,1999800,2499750,3999600,4999500,6249375,7999200,9999000,12498750,19998000,24997500,31246875,39996000,49995000,62493750,99990000,124987500,199980000,249975000,499950000,999900000]</pre>

```python
def solve(a):
    #리스트의 각 값을 다 더한값
    n = 0
    count = 0
    #리스트에서 가장 큰 값
    max_num = 0
    #max_num보다 큰 n의 약수
    divisors = []
    #최종 반환 값
    num = []
    #리스트에서 가장 큰 값 설정
    for i in a:
        if(i>max_num):
            max_num=i
    #리스트의 값을 모두 더한 값 설정
    for i in a:
       n+=i 
    #max_num보다 큰 n의 약수 설정
    for i in range(max_num, n+1):
        if n % i == 0:
            divisors.append(i)
    #해당 약수 값으로 리스트 값을 차례대로 묶을 수 있으면 해당 값을 반환 리스트에 추가
    for i in divisors:
        for j in range(len(a)):
            count+=a[j]
            if(count==i):
                count=0
                if(j==len(a)-1):
                    num.append(i)
            elif count > i:
                count=0
                break
        
    return num
```

```python
solve([9999] * 10**5)
```

```python
import requests
# html 이스케이프 문자 변환을 위한 import문
import html
import string


req_koreatech = requests.get('http://cse.koreatech.ac.kr')
req_nate = requests.get('https://www.nate.com')
req_naver = requests.get('http://www.naver.com')
req_google = requests.get('http://www.google.com')
req_nytimes = requests.get('http://www.nytimes.com')

source_koreatech = req_koreatech.text
source_nate = req_nate.text
source_naver = req_naver.text
source_google = req_google.text
source_nytimes = req_nytimes.text
```

```python
#리스트 중복 제거
def numWord(words):
    a=[]
    for i in words:
        if(not i in a):
            a.append(i)
    return a
```

```python
#리스트 중복 갯수 카운트
def count(word1,word2):
    a=[]
    for i in word1:
        count=0
        for j in word2:
            if i==j:
                count+=1
        a.append(count)
    return a
```

```python
def countWords(source):
    words=[]
    #<script>같은 경우 <script> </script> 사이에 코드나 파일을 삽입하는 코드이므로 해당 명령어 따로 삭제
    while(source.find("<script")+1):
        first_script=source.find("<script")
        last_script=source.find("</script>")
        source=source.replace(source[first_script:last_script+9],'')
    #<script>같은 경우 <style> </style> 사이에 스타일에 관한 명령어를 삽입하는 코드이므로 해당 명령어 따로 삭제
    while(source.find("<style")+1):
        first_style=source.find("<style")
        last_style=source.find("</style>")
        source=source.replace(source[first_style:last_style+8],'')
    #이외 <>들의 명령어 삭제
    while((source.find("<")+1)and(source.find(">")+1)):
        first=source.find("<")
        last=source.find(">")
        source=source.replace(source[first:last+1],'')    
    #html의 이스케이프 문자를 문자로 변환
    source=html.unescape(source)
    #특수문자 제거
    source = ''.join([i for i in source if i not in string.punctuation])
    #공백을 기준으로 리스트화
    words=source.split()
    word1=numWord(words)
    word2=count(word1,words)
    
    dictionary = dict(zip(word1, word2))
    return dictionary
```
